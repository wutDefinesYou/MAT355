### CE 1: Rectangle $ 0 \leq x \leq 3, -2 \leq y \leq 3 $

```octave
n = 4000;

x = zeros(n,1);
y = zeros(n,1);

for i = 1:n
    x(i) = 3*rand();
    y(i) = 5*rand() - 2;
endfor

plot(x,y,"r.")
axis equal
```

### CE 10.1.9: Circle $ (x-3)^2 + (y+1)^2 \leq 9 $

```octave
counter = 0;
while counter < 4000
    a = 6*rand();
    b = 6*rand() - 4;
    if (a - 3)^2 + (b + 1)^2 <= 9
        counter += 1;
        x(counter) = a;
        y(counter) = b;
    endif
endwhile

plot(x,y,"r.")
axis equal
```

### CE 10.1.7 Equilateral Triangle $ (-1,0),(1,0),(0,\sqrt{3}) $

```octave
counter = 0;
while counter < 4000
    a = 2*rand() - 1;
    b = sqrt(3)*rand();
    if (-sqrt(3)*a + sqrt(3) - b) > 0 && (sqrt(3)*a + sqrt(3) - b) > 0
        counter += 1;
        x(counter) = a;
        y(counter) = b;
    endif
endwhile

plot(x,y,"r.")
axis equal
```
